```python
import pandas as pd
import json
```

```python
import json

# load data using Python JSON module
with open("valid_transactions.json", "r") as f:
    data = json.loads(f.read())

# Normalizing data
df = pd.json_normalize(data, record_path=["transactions"])
df.shape[0]
```

```python
df
```

```python
mf = df[df.status == "1"]
```

```python
mf.shape[0]
```

```python
mf = mf.applymap(lambda s: s.lower() if type(s) == str else s)
```

```python
mf.drop_duplicates(subset="address", keep=False, inplace=True)
mf
```

```python
mf = mf.drop(columns=["status", "message"])
mf
```

```python
df_1 = mf.iloc[:50]
df_1
```

```python
for index, row in df_1.iterrows():
    if row["result"][0]["to"] == "" and row["result"][0]["contractAddress"]:
        print(row["address"] + " is a contract address")
```

```python
sf = pd.DataFrame(
    columns=[
        "address",
        "timeStamp",
        "from",
        "to",
        "value",
        "gas",
        "gasPrice",
        "gasUsed",
        "input",
        "txreceipt_status",
        "contractAddress",
    ]
)
for index, row in df_1.iterrows():
    for tx in row["result"]:
        sf = sf.append(
            {
                "address": row["address"],
                "timeStamp": tx["timeStamp"],
                "from": tx["from"],
                "to": tx["to"],
                "value": tx["value"],
                "gas": tx["gas"],
                "gasPrice": tx["gasPrice"],
                "gasUsed": tx["gasUsed"],
                "input": tx["input"],
                "txreceipt_status": tx["txreceipt_status"],
                "contractAddress": tx["contractAddress"],
            },
            ignore_index=True,
        )
sf
```

```python
sf.drop_duplicates(subset="input", keep=False, inplace=True)
sf
```

```python
# storing the data in JSON format
sf.to_json("file.json")
```

```python
sf.to_csv("out.csv")
```

```python
df_2 = sf.iloc[:100]
df_2["address"].unique()
```

```python
caf = pd.DataFrame(
    columns=[
        "contract_create_time",
        "contract_create_txn_fee",
        "contract_create_gas_used",
        "contract_create_gas_price",
        "contract_invoke_first_time",
        "contract_invoke_last_time",
        "contract_active_duration",
        "contract_invocations_total",
        "contract_invocations_unique",
        "contract_invocations_avg_gas",
        "contract_invocations_avg_gas_price",
        "contract_invocations_avg_txn_fee",
        "contract_invocations_total_ether",
        "contract_invocations_avg_ether",
        "contract_invocations_gas",
        "contract_invocations_avg_gas",
    ]
)

for addr in df_2["address"].unique():
    for index, tx in df_2.iterrows():
        feat = {}
        if addr == tx["address"]:
            if tx["to"] == "":
                feat["contract_create_time"] = tx["timeStamp"]
                feat["contract_create_txn_fee"] = tx["gasPrice"]
                feat["contract_create_gas_used"] = tx["gasUsed"]
                feat["contract_create_gas_price"] = tx["gasPrice"]
                feat["contract_invoke_first_time"] = tx["gasPrice"]
                caf = caf.append(feat, ignore_index=True)

caf
```
